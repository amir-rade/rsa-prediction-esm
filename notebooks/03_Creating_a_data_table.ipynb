```python
df = pd.read_parquet('data/processed/dssp_rsa_all_main.parquet')
df["rsa"].isna().sum()
```

```python
import numpy as np
import pandas as pd

dataset = []

grouped = df.groupby(["pdb_id", "chain_id"]) # making multiple ds 

for (pdb_id, chain_id), group in grouped:

    sequence = "".join(group["aa"].tolist())
    rsa_values = group["rsa"].tolist()   # contains NaNs

    dataset.append({
        "pdb_id": pdb_id,
        "chain_id": chain_id,
        "sequence": sequence,
        "rsa": rsa_values
    })

dataset = pd.DataFrame(dataset)

# ensure rsa is numeric; invalid strings -> NaN (kept)
df["rsa"] = pd.to_numeric(df["rsa"], errors="coerce")

# ... after you build dataset_df with rsa as lists ...

def make_mask(rsa_list):
    arr = np.asarray(rsa_list, dtype=float)   # preserves 0.0, turns None/"NA" into nan if present
    return (~np.isnan(arr)).astype(np.int8).tolist() # 

dataset["mask_label"] = dataset["rsa"].apply(make_mask)
```

```python
out_path = "data/processed/rsa_sequence_dataset.parquet"
dataset.to_parquet(out_path, index=False)
```
